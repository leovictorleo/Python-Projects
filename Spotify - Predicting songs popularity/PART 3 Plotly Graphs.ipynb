```python
#import Plotly and Pandas
import plotly as py
import plotly.offline as pyo
from plotly.graph_objs import *
import plotly.graph_objects as go
pyo.offline.init_notebook_mode()

from IPython.display import HTML

import pandas as pd
```

```python
#fetch data
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

data = pd.read_csv('/content/drive/MyDrive/UofT/Machine Learning/ML Group Project/Data Sets/dataset_for_modelling.csv')
```

```python
dataTimePeriod = data[data['year'].isin([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019])]
```

```python
#split dataset by billboard and nonbillboard

billboard = dataTimePeriod[dataTimePeriod["billboard"]==1]
non_billboard = dataTimePeriod[dataTimePeriod["billboard"]==0]
```

```python
billboardGrouped = billboard.groupby(['year'])['danceability','energy','liveness', 'instrumentalness', 'speechiness', 'tempo','acousticness'].mean().reset_index()

billboardGrouped
```

```python
non_billboardGrouped = non_billboard.groupby(['year'])['danceability','energy','liveness', 'instrumentalness', 'speechiness', 'tempo','acousticness'].mean().reset_index()

non_billboardGrouped
```

## 1. Danceability

```python
traceBillboardDanceability = go.Scatter(
                      x = billboardGrouped.year,
                      y = billboardGrouped.danceability,
                      mode = 'lines+markers',
                      line =dict(color='#1539EE'),
                      marker_symbol = 'diamond-dot',
                      marker_line_color = "black",
                      marker_size = 13,
                      marker_line_width = 2,
                      name = 'Billboard Tracks'
                      
)

traceNon_BillboardDanceability = go.Scatter(
                      x = non_billboardGrouped.year,
                      y = non_billboardGrouped.danceability,
                      mode = 'lines+markers',
                      name = 'Non-Billboard Tracks',
                      line =dict(color='#109618'),
                      marker_symbol = 'diamond-dot',
                      marker_line_color = "black",
                      marker_size = 13,
                      marker_line_width = 2
)
```

```python
data_danceability = [traceBillboardDanceability, traceNon_BillboardDanceability]
```

```python
layout_danceability = {'title' : "<b>Danceability by Year<b>",
                       'xaxis' : {'title' : 'Year'},
                       'yaxis' : {'title' : 'Danceability'}}
```

```python
figDance = go.Figure(data = data_danceability, layout = layout_danceability)
figDance.layout.template='plotly_dark'

HTML(figDance.to_html())
```

## Bar Charts

```python
data = pd.read_csv('/content/drive/MyDrive/dataset_for_modelling.csv')
dataTimePeriod = data[data['year'].isin([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019])]
avgData = dataTimePeriod.groupby(['billboard']).mean()
avgData.index = ["Non Billboard","Billboard"]
avgData
```

### EDM + Pop

```python
traceEDM = go.Bar(
    x = avgData.index,
    y = avgData.edm,
    name = 'EDM',
    marker = dict(color='#7316DC')
)

traceDance = go.Bar(
    x = avgData.index,
    y = avgData.dance,
    name = 'Dance',
    marker = dict(color='#0CBF27')
)

traceElectronic = go.Bar(
    x = avgData.index,
    y = avgData.electronic,
    name = 'Electronic',
    marker = dict(color='#EE1515')
)

tracePop = go.Bar(
    x = avgData.index,
    y = avgData['pop'],
    name = 'Pop',
    marker = dict(color='#1539EE')
)
```

```python
dataEDM = [tracePop, traceDance, traceEDM, traceElectronic]
```

```python
layout_EDM = {'title' : "<b>Percentage of Electronic and Pop Music on Billboard<b>", 'barmode' : 'group'}
```

```python
figEDM = go.Figure(data = dataEDM, layout = layout_EDM)
figEDM.layout.template='plotly_dark'

HTML(figEDM.to_html())
```

### Hip Hop

```python
traceRap = go.Bar(
    x = avgData.index,
    y = avgData.rap,
    name = 'Rap',
    marker = dict(color='#1539EE')
)

traceHiphop = go.Bar(
    x = avgData.index,
    y = avgData['hip hop'],
    name = 'Hip Hop',
    marker = dict(color='#0CBF27')
)

traceRandB = go.Bar(
    x = avgData.index,
    y = avgData['r&b'],
    name = 'R&B',
    marker = dict(color='#EE1515')
)
```

```python
dataHiphop = [traceRap, traceHiphop, traceRandB]
```

```python
layoutHipHop = {'title' : "<b>Percentage of Hip Hop, Rap and R&B on Billboard<b>", 'barmode' : 'group'}
```

```python
figHipHop = go.Figure(data = dataHiphop, layout = layoutHipHop)
figHipHop.layout.template='plotly_dark'

HTML(figHipHop.to_html())
```

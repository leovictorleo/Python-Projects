```python
from bs4 import BeautifulSoup 
import pandas as pd 
import requests
from time import sleep
```

```python
def weather_month(year,month,day_start,day_end):
    url = f"http://api.worldweatheronline.com/premium/v1/past-weather.ashx?key=95552a4081f348ce81543624192409&q=San%20Francisco&format=xml&date={year}-{month}-{day_start}&enddate={year}-{month}-{day_end}&tp=1"
    result = requests.get(url)
    result.content
    soup = BeautifulSoup(result.content, "xml")
    soup
    dates = soup.find_all("date")
    hours = soup.find_all("time")
    tempC = soup.find_all("tempC")
    weatherDesc = soup.find_all("weatherDesc")
    precipMM = soup.find_all("precipMM")
    cont = 0
    hour[0].get_text()
    year_list = list()
    month_list = list()
    day_list = list()
    hour_list = list()
    tempC_list = list()
    weatherDesc_list = list()
    precipMM_list = list()
    for i in range(1, len(dates)+1):
        for j in range(cont, cont+24):
            year_list.append(year)
            month_list.append(month)
            day_list.append(i)
            hour_list.append(j%24)
            tempC_list.append(tempC[j].get_text())
            weatherDesc_list.append(weatherDesc[j].get_text())
            precipMM_list.append(precipMM[j].get_text())
            cont += cont
    data_month = {'Year':year_list,'Month':month_list,'Date':day_list,'Hour':hour_list,'tempC':tempC_list,'weatherDesc':weatherDesc_list,'precipMM':precipMM_list}
    data_month_df = pd.DataFrame(data=data_month)
    return(data_month_df)
```
